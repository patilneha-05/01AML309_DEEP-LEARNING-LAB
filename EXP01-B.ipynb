```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD

# -----------------------------
# Model Definition (UNCHANGED)
# -----------------------------
model = Sequential()

# Input layer
model.add(Dense(
    128,
    activation='relu',
    kernel_initializer='he_normal',
    input_shape=(20,)
))

# Hidden layers
model.add(Dense(128, activation='relu', kernel_initializer='he_normal'))  # 2
model.add(Dense(128, activation='relu', kernel_initializer='he_normal'))  # 3
model.add(Dense(64, activation='relu', kernel_initializer='he_normal'))   # 4
model.add(Dense(64, activation='relu', kernel_initializer='he_normal'))   # 5
model.add(Dense(64, activation='relu', kernel_initializer='he_normal'))   # 6
model.add(Dense(32, activation='relu', kernel_initializer='he_normal'))   # 7
model.add(Dense(32, activation='relu', kernel_initializer='he_normal'))   # 8
model.add(Dense(16, activation='relu', kernel_initializer='he_normal'))   # 9

# Output layer
model.add(Dense(1, activation='sigmoid'))

# -----------------------------
# SGD Optimizer (UNCHANGED)
# -----------------------------
sgd = SGD(learning_rate=0.01)

model.compile(
    optimizer=sgd,
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# -----------------------------
# MNIST Dataset (ONLY CHANGE)
# -----------------------------
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# Flatten images
X_train = X_train.reshape(-1, 784)
X_test = X_test.reshape(-1, 784)

# Take first 20 features to match input shape
X_train = X_train[:, :20]
X_test = X_test[:, :20]

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

# Convert labels to binary (0 vs non-zero)
y_train = (y_train > 0).astype(int)
y_test = (y_test > 0).astype(int)

# -----------------------------
# Train Model (UNCHANGED)
# -----------------------------
history = model.fit(
    X_train,
    y_train,
    epochs=25,
    batch_size=32,
    verbose=1
)

# -----------------------------
# Visualization Graphs (UNCHANGED)
# -----------------------------

# Loss Graph
plt.figure()
plt.plot(history.history['loss'])
plt.title('Training Loss (SGD Optimizer)')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()

# Accuracy Graph
plt.figure()
plt.plot(history.history['accuracy'])
plt.title('Training Accuracy (SGD Optimizer)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```
