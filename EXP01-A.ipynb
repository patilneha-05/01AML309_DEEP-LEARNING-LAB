```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# -----------------------------
# Load MNIST Dataset
# -----------------------------
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize pixel values
X_train = X_train / 255.0
X_test = X_test / 255.0

# Flatten 28x28 images into 784 features
X_train = X_train.reshape(-1, 784)
X_test = X_test.reshape(-1, 784)

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# -----------------------------
# Model Definition (10 Layers)
# -----------------------------
model = Sequential()

# Input Layer (Layer 1)
model.add(Dense(
    128,
    activation='relu',
    kernel_initializer='he_normal',
    input_shape=(784,)
))

# Hidden Layers (Layers 2вҖ“9)
model.add(Dense(128, activation='relu', kernel_initializer='he_normal'))  # 2
model.add(Dense(128, activation='relu', kernel_initializer='he_normal'))  # 3
model.add(Dense(64, activation='relu', kernel_initializer='he_normal'))   # 4
model.add(Dense(64, activation='relu', kernel_initializer='he_normal'))   # 5
model.add(Dense(64, activation='relu', kernel_initializer='he_normal'))   # 6
model.add(Dense(32, activation='relu', kernel_initializer='he_normal'))   # 7
model.add(Dense(32, activation='relu', kernel_initializer='he_normal'))   # 8
model.add(Dense(16, activation='relu', kernel_initializer='he_normal'))   # 9

# Output Layer (Layer 10)
model.add(Dense(
    10,
    activation='softmax',
    kernel_initializer='glorot_uniform'
))

# -----------------------------
# Compile Model
# -----------------------------
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# -----------------------------
# Train Model (NO validation)
# -----------------------------
history = model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=32,
    verbose=1
)

# -----------------------------
# Evaluate Model
# -----------------------------
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"\nTest Accuracy: {test_accuracy:.4f}")

# -----------------------------
# Visualization Graphs
# -----------------------------

# Training Loss
plt.figure()
plt.plot(history.history['loss'], label='Training Loss')
plt.title('Loss (He Initialization)')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Training Accuracy
plt.figure()
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.title('Accuracy (He Initialization)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
